```python
# Create a figure with multiple subplots using Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some example data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)

# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Plot on the first subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sin(x)')

# Plot on the second subplot
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cos(x)')

# Plot on the third subplot
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Sin(2x)')

# Plot on the fourth subplot
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Cos(2x)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# â€¢	Use Seaborn to visualize a time series with a trend line.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample time series data (replace this with your actual time series data)
data = {
    'Date': pd.date_range('2023-01-01', '2023-01-10'),
    'Value': [10, 15, 12, 18, 20, 22, 28, 32, 35, 40]
}

df = pd.DataFrame(data)

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a time series plot with a trend line using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Value', data=df, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot with Trend Line')

# Fit a trend line using NumPy's polyfit
trend_coef = np.polyfit(range(len(df)), df['Value'], 1)
trend_line = np.poly1d(trend_coef)
plt.plot(df['Date'], trend_line(range(len(df))), color='red', linestyle='--', label='Trend Line')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
# â€¢	Create a 3D surface plot with a color map.
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x, y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2+y**2))
fig= plt.figure(figsize=(10,5))
ax = fig.add_subplot(111,projection='3d')
surface = ax.plot_surface(x,y,z,cmap='viridis',edgecolor='k',linewidth=0.5)
cbar = fig.colorbar(surface, ax=ax, shrink=0.6, aspect=10)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot with Color Map')
plt.show()
```

```python
# â€¢	Add annotations to specific points on a Matplotlib plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,label='Sin(x)')
annotation_points = [(2, 0.9), (5,0), (8,-0.9)]

for point in annotation_points:
  plt.annotate('Point',
               xy = point,
               xytext = (point[0]+0.5, point[1]+0.5),
               textcoords = 'offset points',
               arrowprops = dict(arrowstyle = '->', connectionstyle = "arc3, rad=.2"))

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Matplotlib Plot with Annonations')
plt.legend()
plt.show()
```

```python
# â€¢	Adjust the spacing and layout of Seaborn subplots.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for plotting
tips = sns.load_dataset("tips")

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Subplot 1: Scatter Plot
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[0, 0])
axes[0, 0].set_title('Scatter Plot')

# Subplot 2: Box Plot
sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[0, 1])
axes[0, 1].set_title('Box Plot')

# Subplot 3: Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips, ax=axes[1, 0])
axes[1, 0].set_title('Violin Plot')

# Subplot 4: Histogram
sns.histplot(tips['total_bill'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Histogram')

# Adjust spacing and layout
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.4, hspace=0.4)

# Show the plot
plt.show()
```

```python
# â€¢	Plot two different datasets with dual y-axes using Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for plotting
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(-0.1 * x) * np.cos(x)

# Create a plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the first dataset on the left y-axis
ax1.plot(x, y1, color='blue', label='Dataset 1')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Dataset 1', color='blue')
ax1.tick_params('y', colors='blue')

# Create a second y-axis on the right
ax2 = ax1.twinx()

# Plot the second dataset on the right y-axis
ax2.plot(x, y2, color='red', label='Dataset 2')
ax2.set_ylabel('Dataset 2', color='red')
ax2.tick_params('y', colors='red')

# Add legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Add title
plt.title('Dual Y-Axes Plot')

# Show the plot
plt.show()
```

```python
# â€¢	Use Seaborn to create a plot with two separate y-axes for different variables.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data for plotting
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(-0.1 * x) * np.cos(x)

# Create a Seaborn plot
sns.lineplot(x=x, y=y1, color='blue', label='Dataset 1')

# Create a Matplotlib Axes object to overlay the second dataset
ax2 = plt.gca().twinx()
ax2.plot(x, y2, color='red', label='Dataset 2')

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Dataset 1', color='blue')
ax2.set_ylabel('Dataset 2', color='red')
plt.title('Seaborn Plot with Dual Y-Axes')

# Add legend
lines, labels = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.gca().legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np

# Sample data for plotting
lons = np.array([10, 20, 30, 40, 50])
lats = np.array([45, 40, 35, 30, 25])
data = np.array([2, 5, 8, 12, 3])

# Create a map with Cartopy
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})

# Plot data on the map
scatter = ax.scatter(lons, lats, c=data, cmap='viridis', s=100, edgecolor='black')

# Add colorbar
cbar = plt.colorbar(scatter, ax=ax, orientation='vertical', pad=0.1)
cbar.set_label('Data Values')

# Add coastlines for better reference
ax.coastlines()

# Add title
plt.title('Data on Map')

# Show the plot
plt.show()
```
